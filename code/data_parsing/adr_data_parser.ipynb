```python
import pandas as pd
```

```python
adr_data = pd.read_excel("Drug_ADR.xlsx")
adr_data
```

```python
smile_data = pd.read_csv("drugbank_smile_data.csv")
smile_data
```

```python
drug_names_with_smile_strings = set(smile_data["name"].tolist())
```

```python
# Ex: get '12, 08' from "12, 12.07.03.015, 12.07.03, 08.02.03.014, 08, 08.02, 08.02.03, 12.07"
def get_adr_categories(adrs_str):
    adrs_list = adrs_str.split(", ")
    adrs_categories = set([x.split(".")[0] for x in adrs_list])
    return ", ".join(list(adrs_categories))
```

```python
name_to_adr_categories_dict = dict()
num_drugs_matched = 0
# 1181 out of 1697 drugs in ADR database match Drugbank SMILES data
# (some matches did not occur due to different spellings of names).
for index, row in adr_data.iterrows():
    curr_name = row['DRUG_NAME']
    if curr_name in drug_names_with_smile_strings:
        curr_adrs = row['ADRECS_ID']
        name_to_adr_categories_dict[curr_name] = get_adr_categories(curr_adrs)
        num_drugs_matched += 1
        
print(num_drugs_matched)
```

```python
smile_data['adrs'] = smile_data['name'].map(name_to_adr_categories_dict)
smile_data
```

```python
# Convert data dictionary to CSV. Write CSV to disk.
smile_data.to_csv("drugbank_smile_data_with_adrs.csv", index=False)
```
